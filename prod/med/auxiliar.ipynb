```python
import pyodbc
import json
import requests
import pandas as pd

import navegadores_json

DRIVER_NAME = 'ODBC Driver 17 for SQL Server'
SERVER_NAME = '.'
DATABASE_NAME = 'ic'

CONNECTION_STRING = (
    f'DRIVER={{{DRIVER_NAME}}};'
    f'SERVER={SERVER_NAME};'
    f'DATABASE={DATABASE_NAME};'
    'Trusted_Connection=yes;'
)

conn = pyodbc.connect(CONNECTION_STRING)
cursor = conn.cursor()
```

```python
sql = """
    SELECT e.Example
    FROM [Endpoints da API] as e
    where e.Id = 470
"""

cursor.execute(sql)
rows = cursor.fetchall()

print(rows)

api = rows[0][0]
```

```python
try:
    resposta = requests.get(api, params={}, timeout=240)
    
    if resposta.status_code == 200:

        print("✅ Requisição bem-sucedida (Status 200).")
        # dados_json = navegadores_json.navegador_id16(resposta.json())
        dados_json = resposta.json()
        print(dados_json)

    else:
        print(f"❌ Erro na requisição: Status Code {resposta.status_code}")
        print(f"Detalhes do erro: {resposta.text}")
        
except requests.exceptions.RequestException as e:
    print(f"❌ Ocorreu um erro ao tentar se conectar à API: {e}")
```

```python
import csv

def txt_para_csv(caminho_txt, caminho_csv=None):
    """
    Lê um arquivo txt onde cada linha está formatada como 'texto1: texto2'
    e converte para um arquivo CSV com duas colunas.
    
    Args:
        caminho_txt (str): Caminho do arquivo txt de entrada
        caminho_csv (str, optional): Caminho do arquivo csv de saída. 
                                     Se não fornecido, usa o mesmo nome do txt com extensão .csv
    
    Returns:
        str: Caminho do arquivo CSV criado
    """
    # Define o caminho do CSV se não foi fornecido
    if caminho_csv is None:
        caminho_csv = caminho_txt.rsplit('.', 1)[0] + '.csv'
    
    # Lista para armazenar os dados
    dados = []
    
    # Lê o arquivo txt
    with open(caminho_txt, 'r', encoding='utf-8') as arquivo_txt:
        for linha in arquivo_txt:
            # Remove espaços em branco no início e fim da linha
            linha = linha.strip()
            
            # Pula linhas vazias
            if not linha:
                continue
            
            # Separa a linha pelo delimitador ": "
            if ': ' in linha:
                partes = linha.split(': ', 1)  # split com limite 1 para dividir apenas na primeira ocorrência
                texto1 = partes[0]
                texto2 = partes[1]
                dados.append([texto1, texto2])
            else:
                # Se a linha não contém o delimitador, você pode optar por ignorá-la
                # ou tratá-la de outra forma
                print(f"Aviso: Linha sem delimitador ': ' encontrada: {linha}")
    
    # Escreve os dados no arquivo CSV
    with open(caminho_csv, 'w', newline='', encoding='utf-8') as arquivo_csv:
        escritor = csv.writer(arquivo_csv)
        
        # Escreve cabeçalho (opcional - remova se não quiser)
        escritor.writerow(['Coluna1', 'Coluna2'])
        
        # Escreve os dados
        escritor.writerows(dados)
    
    print(f"Arquivo CSV criado com sucesso: {caminho_csv}")
    print(f"Total de linhas processadas: {len(dados)}")
    
    return caminho_csv

txt_para_csv("Relatório dos endpoints falhos.txt", "Relatório dos endpoints falhos..csv")
```
